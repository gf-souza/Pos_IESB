```python
#importando a biblioteca
import pandas as pd

#ler o arquivo e criar DataFrame

df = pd.read_excel ('201710_EmissaoPassagens_SCDP.xlsx', skiprows=3, skipfooter=3)
```

```python
#verificando os nomes da colunas

df.columns
```

```python
#substituindo os nomes das colunas

df=df.rename(columns={
    ' Código do órgão superior':'Código do órgão superior',
    '                               Nome do órgão superior                                ':'Nome do órgão superior',
    ' Código do órgão solicitante da viagem':'Código do órgão solicitante da viagem',
    '                                      Nome do órgão solicitante da viagem                                       ':'Nome do órgão solicitante da viagem',
    '    N. PCDP    ':'N. PCDP',
    ' N. Reserva/Localizador':'N. Reserva/Localizador',
       '  Data Emissão Bilhete   ':'Data Emissão Bilhete',
    '    Data Embarque    ':'Data Embarque',
       ' Valor Tarifa Comercial':'Valor Tarifa Comercial',
    ' Percentual Desconto Aplicado':'Percentual Desconto Aplicado',
       ' Valor Tarifa Governo':'Valor Tarifa Governo',
    ' Valor Tarifa Embarque':'Valor Tarifa Embarque',
    ' Valor Bilhete':'Valor Bilhete',
       '           Companhia Aérea            ':'Companhia Aérea',
    ' Classe Tarifária Bilhete':'Classe Tarifária Bilhete',
       '                                       Regra Tarifária                                       ':'Regra Tarifária',
       ' No Show':'No Show', 
    ' Remarcado':'Remarcado',
    ' Cancelado':'Cancelado', 
    ' Valor Multas':'Valor Multas',
       ' Valor Reembolso':'Valor Reembolso', 
    ' Diferença de Tarifa':'Diferença de Tarifa',
    ' Situação Final Bilhete':'Situação Final Bilhete'})
```

```python
# verificando os tipos de dados da coluna
df.info()
```

```python
#verificando os 10 voos com maior valor de bilhete

df.nlargest(10,'Valor Bilhete' )
```

```python
#verificando quais foram as companhias aereas utilizadas

df['Companhia Aérea'].value_counts()
```

```python
df.groupby(['Companhia Aérea'])['Valor Bilhete'].sum()
```

```python
#criando uma nova coluna com o valor de bilhete em milhoes 
df['Bilhete_milhao'] = df['Valor Bilhete']/1000000
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df.groupby(['Companhia Aérea'])['Bilhete_milhao'].sum().plot.barh()
plt.title('Gasto total por Companhia Aérea')
plt.xlabel('Em milhões de R$')
```

```python
#verificando o numero de No Show

df['No Show'].value_counts(normalize=True).plot.pie()
plt.ylabel('')
plt.title('Percentagem de No Show')

```
